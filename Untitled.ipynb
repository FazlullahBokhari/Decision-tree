```python
# Import the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
# Import the dataset
dataset = pd.read_csv('DT.csv - DT.csv.csv')
```

```python
dataset
```

```python
input1 = dataset.drop('salare>100k',axis=1)
input1
```

```python
target = dataset['salare>100k']
target
```

```python
from sklearn.preprocessing import LabelEncoder
comp = LabelEncoder()
job = LabelEncoder()
deg = LabelEncoder()
```

```python
input1['comp_new'] = comp.fit_transform(input1['company'])
```

```python
input1['job_new'] = job.fit_transform(input1['job'])
```

```python
input1['deg_new'] = deg.fit_transform(input1['degree'])
```

```python
input1
```

```python
input2 = input1.drop(columns=['company','job','degree'])
input2
```

```python
# Fit Decision Tree Classifier to the Training set
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(input2, target) 
```

```python
score = model.score(input2, target)
score
```

```python
model.predict([[2,2,0]])
```
